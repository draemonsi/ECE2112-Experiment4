```python
# Import the following libraries that will be used in the manipulation and interpretation of data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read the csv file
board = pd.read_csv('board2.csv')
```

```python
board
```

```python
# Store the read csv file in a DataFrame
board_results = pd.DataFrame(board)
```

```python
board_results
```

#### Create the following data frames based on the format provided:##### Example: Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as Visayas

```python
# First is to set what columns will be shown in the table, 
# next is what rows will be shown depending on the Hometown if they're
# from Visayas and if their Math grade is less than 70.
# using conditional indexing

board_results[['Name','Gender','Track','Math']][(board_results['Hometown'] == 'Visayas') & (board_results['Math']<70)]
```

##### Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as##### Instrumentation and hometown Luzon

```python
# Set columns that will be shown in the table
# Determine the rows that will be selected depending on the score they got from
# electronics, and if their track is instrumentation and they are from Luzon using
# conditional indexing

board_results[['Name','GEAS','Electronics']][(board_results['Electronics'] > 70) & 
(board_results['Track'] == 'Instrumentation') & 
(board_results['Hometown'] == 'Luzon')]
```

##### Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is##### constant as Mindanao and gender Female

```python
# Create a new column for the Average scores of board takers
# By getting the sum of all the subjects and then dividing the sum by 4, we can get the average.
board_results['Average'] = (board_results['Math'] + board_results['Electronics'] + board_results['GEAS'] + board_results['Communication']) / 4

# Determine which columns are row will be shown in the table.
# Use conditional indexing for the rows since we need the average to be >= 55, with their hometown being in Mindanao and gender 
# being female.
board_results[['Name', 'Track', 'Electronics','Average']][(board_results['Average'] >= 55) & (board_results['Hometown'] == 'Mindanao') &
(board_results['Gender'] == 'Female')]
```

```python
# List of subjects to plot
subjects = ["Math", "Electronics", "GEAS", "Communication"]

# Create a boxplot for each subject, grouped by "Track"
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 6))

# To create the parameters for each boxplot
for idx, subject in enumerate(subjects):
    board_results.boxplot(column=subject, by='Track', ax=axes[idx])
    axes[idx].set_title(f'{subject} Scores by Track')
    axes[idx].set_xlabel('Track')
    axes[idx].set_ylabel(subject)

plt.suptitle('')  # Suppress the default title
plt.tight_layout()
plt.show()
```

##### Data interpretation: 
By looking at the boxplot above, we can see that in the Math scores, microelectronics has the highest percentage of students that got a score above 87. However, we can see that the mean of students in the microelectronics track in the Electronics section of the board exam has a lower value compared to those in the other tracks. In the GEAS section, we can see that no students in the Instrumentation track have scored lower than 55, in comparison to the other tracks. However, in the communication section, no student from the instrumentation track has scored above 80 in comparison to the other tracks and the instrumentation track also has the lowest score mean in the communication section.

```python
subjects = ["Math", "Electronics", "GEAS", "Communication"]
tracks = board_results["Track"].unique()

# Compute the average scores for each track and subject
avg_scores = board_results.groupby("Track")[subjects].mean()

# Set up the bar graph
bar_width = 0.2
index = np.arange(len(tracks))

# Create subplots for each subject
fig, ax = plt.subplots(figsize=(10, 6))

# For each subject, create a bar
for i, subject in enumerate(subjects):
    ax.bar(index + i * bar_width, avg_scores[subject], bar_width, label=subject)

# Set the x-axis labels and positions
ax.set_xlabel('Track')
ax.set_ylabel('Average Scores')
ax.set_title('Average Scores by Track and Subject')
ax.set_xticks(index + bar_width * 1.5)  # Adjust position for labels
ax.set_xticklabels(tracks)

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

##### Data interpretation: 
By analyzing the bar graph above between the relationship of Track and scores in subjects, we can clearly see that the Microelectronics track excels in Math and not so much in electronics in comparison to other tracks. On the other hand, instrumentation excels in the GEAS section of the board exam and scored averagely in the other sections. However for the communication track, we can see that they didnt excel in anything particular but they scored averagely in every section.

```python
track_gender_counts = board_results.groupby(['Track', 'Gender']).size().unstack()

# Set up the figure for multiple pie charts
fig, axes = plt.subplots(nrows=1, ncols=len(track_gender_counts), figsize=(18, 6))

# Plotting a pie chart for each track
for idx, track in enumerate(track_gender_counts.index):
    axes[idx].pie(track_gender_counts.loc[track], labels=track_gender_counts.columns, autopct='%1.1f%%', startangle=90)
    axes[idx].set_title(f'Gender Distribution in {track}')

# Adjust layout
plt.tight_layout()
plt.show()
```

##### Data interpretation: 
The communication track has an equal distribution of male and female students. Instrumentation track is predominantly male while Microelectronics is predominantly female.

```python
# Set up the plot
plt.figure(figsize=(10, 6))

# Scatter plot: Math vs. Electronics, colored by Hometown
sns.scatterplot(data=board_results, x="Math", y="Electronics", hue="Hometown", palette="Set1", s=100)

# Add labels and title
plt.title('Scatter Plot of Math vs Electronics Scores by Hometown')
plt.xlabel('Math Scores')
plt.ylabel('Electronics Scores')

# Show the plot
plt.show()
```

##### Data interpretation: 
We can analyze from the scatterplot above that more than half of the students who took the board exam from Visayas excelled in Math and had a significant difference in scores from students in Luzon and Mindanao, however, none of the students from Visayas passed in the Electronics section of the board exam. In Luzon and Mindanao, less than half of the students who took the exam passed both Math and electronics.

```python
# Set up the plot
plt.figure(figsize=(10, 6))

# Scatter plot: Math vs. Electronics, colored by Track
sns.scatterplot(data=board_results, x="Math", y="Electronics", hue="Track", palette="Set1", s=100)

# Add labels and title
plt.title('Scatter Plot of Math vs Electronics Scores by Track')
plt.xlabel('Math Scores')
plt.ylabel('Electronics Scores')

# Show the plot
plt.show()
```

#### Data interpretation
We can see in this other version of the scatterplot, which groups the colors by track, that those whose track is microelectronics excelled in mathematics but the majority of them have failed in electronics further supporting the data from the bar graph. We can also interpret from this scatterplot that none of the board takers have >80 score in both Electronics and Math score, it either they excel in Math but do bad in Electronics.

```python
# Set up the plot
plt.figure(figsize=(10, 6))

# Scatter plot: GEAS vs. Communication, colored by Track
sns.scatterplot(data=board_results, x="GEAS", y="Communication", hue="Track", palette="Set1", s=100)

# Add labels and title
plt.title('Scatter Plot of GEAS vs Communication Scores by Track')
plt.xlabel('GEAS Scores')
plt.ylabel('Communication Scores')

# Show the plot
plt.show()
```

#### Data interpretation
Scores in GEAS and Communication seem independent of each other. Students who performed well in one section do not necessarily excel in the other, across all tracks.
No clear pattern of strong performance across both subjects was observed, which indicates variability in student performance based on track and possibly other factors.
